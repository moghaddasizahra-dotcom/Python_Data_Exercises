```python
#Question: "Write the code to create a new column 'grade' where marks >= 85 are 'A', 70-84 are 'B', 60-69 are 'C', and below 60 are 'D'."

def get_grade (mark):
  if mark >= 85:
    return 'A'
  elif mark >= 70:
    return 'B'
  elif mark >= 60:
    return 'C'
  else:
    return 'D'

df['grade'] = df['mark'].apply(get_grade)

print(df)
```

```python
import pandas as pd

sorted_df = df.sort_values(by='mark', ascending=False)

print (sorted_df)
```

**Exporting Data**

```python
import pandas as pd

sorted_df = df.sort_values(by='mark', ascending=False)

print(sorted_df)

sorted_df.to_csv('sorted_student.csv', index=False)
```

**Visualising the result**

```python
from matplotlib import pyplot as plt
import seaborn as sns
sorted_df.groupby('grade').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)

plt.xlabel('Number of Students')
plt.ylabel('Grade')
plt.title('Number of Students per Grade')
plt.show()
```
